- **The question**
    - <code style="background:darkblue;color:black">How much land suitable for cultivation (CL) do we actually have in the Earth per human being?</code>
    
This mean, if we divide the quantity of Land suitable for cultivation and grazing, with the total world population, how much land do we have?

For this, we defined tree main indicators:

|Indicator|Meaning|Equation|
|:----|:-----|:-----|
| 1 *Cultivables Lands (CL)*|Lands **already** intended to be <br>planted and doing grazing.|\begin{equation*}CL=AL+PC\end{equation*}|
| 2 *Permaculture Indicator (Perm_ind)*| Minimun amount of land needed <br>per family. **It's a fix value.**|\begin{equation*}Perm ind = 1 Ha \end{equation*}|
|3 *Hectares per Family (HpF)*| Calculated CL land per family |\begin{equation*}HpF = \frac{CL}{Family}\end{equation*}|
